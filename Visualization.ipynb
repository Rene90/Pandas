```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bb = data.DataReader(name = "GOOG", data_source = "yahoo", start = "2004-01-01", end = "2016-12-12")
bb.head()
```

```python
bb.plot(y = "High")
```

```python
bb["Close"].plot()
```

```python
#modificando la estetica
plt.style.available
```

```python
plt.style.use("ggplot")
bb.plot(y="Close")
```

```python
#bar charts
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind = "barh")
```

```python
# pie charts
bb["Close"].mean()
```

```python
def rank(stock):
    if stock >= 251.57074582479834:
        return "Above average"
    else:
        return "Below average"
```

```python
bb["Close"].apply(rank).value_counts().plot(kind="pie")
```

```python
#histogramas
def custom_round(stock):
    return int(stock /100.0)*100
```

```python
bb["High"].apply(custom_round).value_counts()
```

```python
bb["High"].apply(custom_round).nunique()
```

```python
bb["High"].apply(custom_round).plot(kind = "hist", bins = 9)
```
